```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans

```

```python
def plot_TSNE(df):
    ax = df.plot.scatter('X','Y')
    ax.set_xlabel('X', fontsize = 15)
    ax.set_ylabel('Y', fontsize = 15)
    ax.set_title('2 component TSNe', fontsize = 20)
    return ax

def TSNE_n_plot(df, PCA_comp, TSNE_comp, perplexity, pvw_h, iter_TSNE = 1000):
    #RECIBE ENTRADAS POR COLUMNAS
    if pvw_h == True:
        cols2drop = ['year','day','hour']
    else:
        cols2drop = ['hidrology','week']
    df = df.drop(columns=cols2drop)
    array_datos = df.to_numpy()
    array_datos = array_datos.transpose()
    array_datos = StandardScaler().fit_transform(array_datos)
    pca = PCA(n_components = PCA_comp)
    ppalComponents = pca.fit_transform(array_datos)
    cols = ['PCA Component ' + str(i) for i in range(1, PCA_comp + 1)]
    pca_df = pd.DataFrame(data = ppalComponents, columns = cols)
    tsne_data = pca_df.to_numpy()
    X_tsne = TSNE(n_components = TSNE_comp, perplexity = perplexity, verbose = 1, n_iter = iter_TSNE).fit_transform(tsne_data)
    df_X_tsne = pd.DataFrame(data = X_tsne, columns = ['X','Y'])
    gens = pd.DataFrame(data = df.columns, columns = ['Generadores'])
    df_final = pd.concat([df_X_tsne, gens], axis = 1)
    plot_TSNE(df_final)
    return df_final
```

```python
def kmeans_plot(df, n_clusters):
    data = df.loc[:,['X','Y']].to_numpy()
    kmeans = KMeans(n_clusters = n_clusters, random_state = 0).fit(data)
    data_labels = ['Cluster ' + str(i) for i in kmeans.labels_]
    output_df = pd.concat([df, pd.DataFrame(data_labels, columns = ['Cluster'])], axis = 1)
    labels = set(data_labels)
    colors = ['b','g','r','c','m','y','k']
    fig = plt.figure(figsize = (8,8))
    ax2 = fig.add_subplot(1,1,1) 
    ax2.set_xlabel('Principal Component 1', fontsize = 15)
    ax2.set_ylabel('Principal Component 2', fontsize = 15)
    ax2.set_title('2 components TSNE', fontsize = 20)
    for idx, label in enumerate(labels):
        indices2Keep = output_df['Cluster'] == label
        ax2.scatter(output_df.loc[indices2Keep,'X'], output_df.loc[indices2Keep,'Y'], c = colors[idx], s = 50)
    return output_df
```

```python
# df_datos_solar = pd.read_csv('perfiles.csv',encoding = "cp1252")
# df_datos_wind = pd.read_csv('perfiles_wind.csv',encoding = "utf-8")
df_datos_hidro_N = pd.read_csv('inflows_normalized_N.csv')
#df_datos_hidro_S = pd.read_csv('inflows_normalized_S.csv')
```

```python
df_datos_hidro_N
```

```python
hydro_N = TSNE_n_plot(df_datos_hidro_N, PCA_comp = 6, TSNE_comp = 2, perplexity = 12, pvw_h = False, iter_TSNE = 10000)
```

```python
hydro_S = TSNE_n_plot(df_datos_hidro_S, PCA_comp = 20, TSNE_comp = 2, perplexity = 40, pvw_h = False, iter_TSNE = 1000)
```

```python
kmeans_hydro_N = kmeans_plot(hydro_N,5)
```

```python
kmeans_hydro_S = kmeans_plot(hydro_S,5)
```

```python
kmeans_hydro_N.to_csv('cluster_hydro_N.csv')
kmeans_hydro_S.to_csv('cluster_hydro_S.csv')
```

```python
fig_hydro_S = px.scatter(kmeans_hydro_S, x = "X", y = "Y", color = "Cluster", hover_data = ["Generadores"])
fig_hydro_S.show()
```

```python
inflows_hidro_N = df_datos_hidro_N.melt(id_vars = ['week','hidrology'], var_name = "Central", value_name = "Afluente")
inflows_hidro_N = pd.pivot_table(inflows_hidro_N, values = "Afluente", index = ["Central","week"], aggfunc = "mean").reset_index()
inflows_hidro_N = inflows_hidro_N.merge(kmeans_hydro_N, left_on = 'Central', right_on = 'Generadores').drop(columns = ["X","Y","Generadores"])
fig4 = px.line(inflows_hidro_N, x = "week", y = "Afluente", color = "Cluster", line_group = "Central", hover_name = "Central")
fig4.show()
```

```python
inflows_hidro_S = df_datos_hidro_S.melt(id_vars = ['week','hidrology'], var_name = "Central", value_name = "Afluente")
inflows_hidro_S = pd.pivot_table(inflows_hidro_S, values = "Afluente", index = ["Central","week"], aggfunc = "mean").reset_index()
inflows_hidro_S = inflows_hidro_S.merge(kmeans_hydro_S, left_on = 'Central', right_on = 'Generadores').drop(columns = ["X","Y","Generadores"])
fig5 = px.line(inflows_hidro_S, x = "week", y = "Afluente", color = "Cluster", line_group = "Central", hover_name = "Central")
fig5.show()
```
