```python
import pandas as pd
import os
#cwd = os.getcwd()
#print(cwd)
```

```python
gen = pd.read_csv('gen.csv',encoding = "ANSI") 
gen = gen[gen.type == 'ror']
hydro = pd.read_csv('hydro.csv',encoding = "utf-8")
inflows = pd.read_csv('inflows.csv',encoding = "utf-8")
```

```python
gen = gen.drop(columns = ["IFOR","category","type","subsystem"])

hydro = hydro.drop(columns = ["vmax","vmin","filtrated","turbined","spilled"])
```

```python
inflows
```

```python
inflows = inflows.melt(id_vars = ["week","hidrology"], var_name = "Central", value_name = "Afluente")
```

```python
inflows
```

```python
df = inflows.merge(gen, left_on = "Central",right_on = "name")
```

```python
df = df.merge(hydro, left_on = "Central", right_on = "name")
df
```

```python
df = df.drop(columns = ["name_x","name_y"])
df
```

```python
df['P_out'] = df.apply(lambda x: min(1,x['Afluente']*x['rend']/x['capacity']), axis = 1)
df = df.drop(columns = ['Afluente','capacity','rend'])
df
```

```python
df2 = df.pivot_table(index=['week','hidrology'], columns = 'Central',values = 'P_out').reset_index()
df2
```

```python
df2.to_csv('inflows_normalized_N.csv', index=False)
```
